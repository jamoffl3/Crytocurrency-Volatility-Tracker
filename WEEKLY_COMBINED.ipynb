**WEEK-3:**  The Cleaning Sprint. Handling Missing Values (Imputation
strategies) and Outlier Detection (Boxplots/Z-score).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

print("WEEK 3 PRACTICAL : DATA CLEANING SPRINT\n")

print("Loading Datasets...\n")

# Corrected filename and separator for crypto_df
crypto_df = pd.read_csv(
    r"/content/Bitcoin_12_4_2025-2_5_2026_historical_data_coinmarketcap.csv", sep=';'
)

# Using sentiment_df for consistency; parameters are already correct
sentiment_df = pd.read_csv(
    r"/content/crypto_10k_tweets_(2021_2022Nov).csv",
    engine="python",
    encoding="latin1",
    on_bad_lines="skip"
)

print("Datasets Loaded Successfully\n")

print("Checking Missing Values...\n")

print("Crypto Dataset Missing Values:")
print(crypto_df.isnull().sum(), "\n")

print("Sentiment Tweets Dataset Missing Values:")
print(sentiment_df.isnull().sum(), "\n")

print("Handling Missing Values (Imputation)...\n")

# Convert datetime columns in crypto_df
datetime_cols_crypto = ['timeOpen', 'timeClose', 'timeHigh', 'timeLow', 'timestamp']
for col in datetime_cols_crypto:
    crypto_df[col] = pd.to_datetime(crypto_df[col], errors='coerce')

# The crypto_df's numeric columns ('open', 'high', 'low', 'close', 'volume', 'marketCap', 'circulatingSupply')
# are already correctly typed and do not need the string replacement logic for object columns.
# Selecting numeric columns ONLY from crypto_df for outlier detection and imputation
crypto_numeric = crypto_df.select_dtypes(include=np.number)

# Drop columns that are fully NaN
crypto_numeric = crypto_numeric.dropna(axis=1, how="all")

# Mean imputation
# Apply imputation only to the numeric subset and assign back
crypto_df[crypto_numeric.columns] = crypto_numeric.fillna(
    crypto_numeric.mean()
)

print("Missing values in Crypto Dataset handled using Mean Imputation\n")

# Handle sentiment_df dataset: check for 'Content' column instead of 'tweet'
if "Content" in sentiment_df.columns:
    sentiment_df = sentiment_df.dropna(subset=["Content"])
else:
    sentiment_df = sentiment_df.dropna()

print("Missing values in Sentiment Dataset handled by Row Deletion\n")

print("Outlier Detection using Boxplot...\n")

if crypto_numeric.shape[1] == 0:
    print("No numeric columns available for boxplot.\n")
else:
    plt.figure(figsize=(10, 6))
    crypto_numeric.boxplot()
    plt.title("Boxplot for Crypto Dataset (Outlier Detection)")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

print("Outlier Detection using Z-Score...\n")

z_scores = np.abs(stats.zscore(crypto_numeric, nan_policy="omit"))
outliers = (z_scores > 3).any(axis=1)

print("Number of Outlier Rows Detected:", outliers.sum(), "\n")

# Remove outliers
crypto_df_cleaned = crypto_df.loc[~outliers]

print("Outliers Removed Successfully")
print("Original Crypto Dataset Shape:", crypto_df.shape)
print("Cleaned Crypto Dataset Shape:", crypto_df_cleaned.shape, "\n")

print("Week 3 Data Cleaning Completed Successfully.")
```

```python
# Install mplfinance if not already installed
!pip install mplfinance

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import mplfinance as mpf
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

finance_df = pd.read_csv(
    "/content/Bitcoin_12_4_2025-2_5_2026_historical_data_coinmarketcap.csv",
    sep=";"
)
tweets_df = pd.read_csv(
    "/content/crypto_10k_tweets_(2021_2022Nov).csv",
    engine="python",
    encoding="latin1",
    on_bad_lines="skip"
)
finance_df["timeOpen"] = pd.to_datetime(finance_df["timeOpen"], errors="coerce")
numeric_cols = [
    "open", "high", "low", "close",
    "volume", "marketCap", "circulatingSupply"
]
for col in numeric_cols:
    finance_df[col] = (
        finance_df[col]
        .astype(str)
        .str.replace(",", "", regex=True)
        .str.replace("$", "", regex=True)
        .str.replace("%", "", regex=True)
    )
    finance_df[col] = pd.to_numeric(finance_df[col], errors="coerce")

finance_numeric = finance_df[numeric_cols]
plt.figure(figsize=(6,4))
finance_df["close"].hist(bins=30)
plt.title("Distribution of Bitcoin Closing Price")
plt.xlabel("Closing Price")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6,4))
finance_df["volume"].hist(bins=30)
plt.title("Distribution of Trading Volume")
plt.xlabel("Volume")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6,4))
finance_numeric.boxplot()
plt.title("Boxplot of Bitcoin Numeric Features")
plt.xticks(rotation=45)
plt.show()

tweets_df["Date"] = pd.to_datetime(tweets_df["Date"], errors="coerce")
tweets_df["tweet_length"] = tweets_df["Content"].astype(str).apply(len)

plt.figure(figsize=(6,4))
tweets_df["tweet_length"].hist(bins=30)
plt.title("Distribution of Tweet Lengths")
plt.xlabel("Tweet Length")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6,4))
plt.scatter(finance_df["close"], finance_df["volume"], alpha=0.6)
plt.title("Bitcoin Closing Price vs Trading Volume")
plt.xlabel("Closing Price")
plt.ylabel("Volume")
plt.show()

plt.figure(figsize=(8,4))
plt.plot(finance_df["timeOpen"], finance_df["close"])
plt.title("Bitcoin Closing Price Over Time")
plt.xlabel("Date")
plt.ylabel("Closing Price")
plt.show()

corr_matrix = finance_numeric.corr()
print("Correlation Matrix:\n")
print(corr_matrix)

print("\nCorrelation between Closing Price and Volume:")
print(corr_matrix.loc["close", "volume"])


candle_df = finance_df[
    ["timeOpen", "open", "high", "low", "close"]
].copy()
candle_df.rename(
    columns={
        "timeOpen": "Date",
        "open": "Open",
        "high": "High",
        "low": "Low",
        "close": "Close",
    },
    inplace=True,
)
candle_df["Date"] = pd.to_datetime(candle_df["Date"])
candle_df.set_index("Date", inplace=True)

candle_df.sort_index(inplace=True)

mpf.plot(
    candle_df,
    type="candle",
    style="yahoo",
    title="Bitcoin Candlestick Chart",
    ylabel="Price",
    xlabel="Date",
    volume=False,
    figsize=(10, 5),
)

# Re-defining numeric_cols as it might not be in the current execution scope
numeric_cols = [
    "open", "high", "low", "close",
    "volume", "marketCap", "circulatingSupply"
]

# Re-processing finance_df to create finance_numeric as it was not defined
# Assuming finance_df is already loaded and timeOpen is converted from previous steps
for col in numeric_cols:
    finance_df[col] = (
        finance_df[col]
        .astype(str)
        .str.replace(",", "", regex=True)
        .str.replace("$", "", regex=True)
        .str.replace("%", "", regex=True)
    )
    finance_df[col] = pd.to_numeric(finance_df[col], errors="coerce")

finance_numeric = finance_df[numeric_cols]

corr_matrix = finance_numeric.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap of Bitcoin Financial Features")
plt.show()

finance_df = finance_df.sort_values(by='timeOpen')
finance_df['Price_Change'] = finance_df['close'].diff()
finance_df['Price_Movement'] = np.where(finance_df['Price_Change'] > 0, 'Up', 'Down')

print(finance_df[['timeOpen', 'close', 'Price_Change', 'Price_Movement']].head())


# Drop the first row which will have NaN for Price_Change
finance_df_classified = finance_df.dropna(subset=['Price_Change']).copy()

# Define features (X) and target (y)
# Using relevant numerical features for classification
X = finance_df_classified[numeric_cols].drop(columns=['close'])
y = finance_df_classified['Price_Movement']

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")

# Initialize and train a RandomForestClassifier model
model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Print the confusion matrix
print("Confusion Matrix:")
print(conf_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(
    conf_matrix,
    annot=True,
    fmt="d",
    cmap="Blues",
    cbar=False,
    xticklabels=model.classes_,
    yticklabels=model.classes_
)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix for Bitcoin Price Movement Classification")
plt.show()
```

**WEEK-6:** Feature Engineering. Normalization, Scaling, and Encoding
categorical variables. Fixing skewness.

```python
print("WEEK 6 PRACTICAL : FEATURE ENGINEERING (FULL VERSION)\n")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import skew, kurtosis
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder

# -------------------------------
# 0. DEFINE NUMERIC COLUMNS
# -------------------------------

numeric_cols = [
    "open", "high", "low", "close",
    "volume", "marketCap", "circulatingSupply"
]

# -------------------------------
# 1. SKEWNESS & KURTOSIS ANALYSIS
# -------------------------------

print("Calculating Skewness and Kurtosis...\n")

skewness_vals = finance_df[numeric_cols].apply(skew)
kurtosis_vals = finance_df[numeric_cols].apply(kurtosis)

print("Skewness:\n", skewness_vals, "\n")
print("Kurtosis:\n", kurtosis_vals, "\n")

# -------------------------------
# 2. DISTRIBUTION PLOTS
# -------------------------------

print("Plotting distributions...\n")

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(finance_df[col], kde=True)
    plt.title(f"{col}\nSkew={skewness_vals[col]:.2f}, Kurt={kurtosis_vals[col]:.2f}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.show()

# -------------------------------
# 3. BOXPLOT (OUTLIERS)
# -------------------------------

plt.figure(figsize=(10,6))
sns.boxplot(data=finance_df[numeric_cols])
plt.title("Boxplot Before Transformation")
plt.xticks(rotation=45)
plt.show()

# -------------------------------
# 4. SKEWNESS BAR GRAPH
# -------------------------------

plt.figure(figsize=(8,5))
skewness_vals.sort_values().plot(kind='barh')
plt.title("Feature Skewness")
plt.xlabel("Skewness")
plt.show()

# -------------------------------
# 5. KURTOSIS BAR GRAPH
# -------------------------------

plt.figure(figsize=(8,5))
kurtosis_vals.sort_values().plot(kind='barh')
plt.title("Feature Kurtosis")
plt.xlabel("Kurtosis")
plt.show()

# -------------------------------
# 6. HANDLE SKEWNESS (LOG TRANSFORM)
# -------------------------------

print("\nApplying Log Transformation...\n")

skewed_cols = skewness_vals[abs(skewness_vals) > 1].index

for col in skewed_cols:
    # Ensure values are positive before log transform, replace non-positives with NaN if they occur
    finance_df[col] = np.log1p(finance_df[col].apply(lambda x: x if x > -1 else np.nan))

print("Transformed Columns:", list(skewed_cols))

# -------------------------------
# 7. CHECK AGAIN AFTER TRANSFORM
# -------------------------------

skewness_after = finance_df[numeric_cols].apply(skew)
kurtosis_after = finance_df[numeric_cols].apply(kurtosis)

print("\nSkewness After:\n", skewness_after, "\n")
print("Kurtosis After:\n", kurtosis_after, "\n")

for col in skewed_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(finance_df[col], kde=True)
    plt.title(f"{col} AFTER Transformation\nSkew={skewness_after[col]:.2f}")
    plt.show()

# -------------------------------
# 8. FEATURE CREATION
# -------------------------------

print("\nCreating New Features...\n")

finance_df["High_Low_Range"] = finance_df["high"] - finance_df["low"]
finance_df["Open_Close_Diff"] = finance_df["open"] - finance_df["close"]

# Handle division by zero or NaN for 'marketCap' to prevent 'inf' or further 'NaN'
finance_df["Volume_to_MarketCap"] = finance_df["volume"] / finance_df["marketCap"]

print(finance_df[[
    "High_Low_Range",
    "Open_Close_Diff",
    "Volume_to_MarketCap"
]].head())

# -------------------------------
# Handle NaN and Infinite values before Scaling
# -------------------------------
# Define all features that will be scaled, including newly created ones
features_to_scale = numeric_cols + [
    "High_Low_Range",
    "Open_Close_Diff",
    "Volume_to_MarketCap"
]

# Replace infinite values with NaN
finance_df.replace([np.inf, -np.inf], np.nan, inplace=True)

# Impute remaining NaN values with the mean of their respective columns
# Ensure that the mean is calculated only from non-NaN values
for col in features_to_scale:
    if finance_df[col].isnull().any():
        finance_df[col].fillna(finance_df[col].mean(), inplace=True)

print("\nNaN and Infinite values handled before scaling.\n")


# -------------------------------
# 9. SCALING (STANDARDIZATION)
# -------------------------------

print("\nApplying Standard Scaling...\n")

scaler = StandardScaler()
finance_df[features_to_scale] = scaler.fit_transform(finance_df[features_to_scale])

# -------------------------------
# 10. NORMALIZATION (MIN-MAX)
# -------------------------------

print("Applying Min-Max Normalization...\n")

minmax = MinMaxScaler()
finance_df[features_to_scale] = minmax.fit_transform(finance_df[features_to_scale])

# -------------------------------
# 11. ENCODING TARGET VARIABLE
# -------------------------------

print("Encoding Price Movement...\n")

le = LabelEncoder()
finance_df["Price_Movement_Encoded"] = le.fit_transform(finance_df["Price_Movement"])

print(finance_df[["Price_Movement", "Price_Movement_Encoded"]].head())

# -------------------------------
# 12. FINAL FEATURE SET
# -------------------------------

X = finance_df[features_to_scale]
y = finance_df["Price_Movement_Encoded"]

print("\nFinal Feature Shape:", X.shape)
print("Target Shape:", y.shape)

print("\n✅ Week 6 Feature Engineering Completed Successfully.")
```

**WEEK-7:**  7
Dimensionality Reduction. Implementing Feature Selection and
PCA. Visualizing variance

```python
print("WEEK 7 PRACTICAL : DIMENSIONALITY REDUCTION (TOP FEATURES)\n")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA

# -------------------------------
# 0. INPUT FROM WEEK 6
# -------------------------------

print("Original Feature Count:", X.shape[1])

# -------------------------------
# 1. FEATURE SELECTION (RFE)
# -------------------------------

print("\nApplying RFE to select top features...\n")

model = RandomForestClassifier(random_state=42)

# 🔥 SET NUMBER OF FEATURES (CHANGE HERE if needed)
n_features_to_select = min(12, X.shape[1])  # ensures no error if <12 features

rfe = RFE(estimator=model, n_features_to_select=n_features_to_select)
X_rfe = rfe.fit_transform(X, y)

selected_features = X.columns[rfe.support_]

print(f"Selected Top {len(selected_features)} Features:\n")

for i, feature in enumerate(selected_features, 1):
    print(f"{i}. {feature}")


# -------------------------------
# 2. FEATURE IMPORTANCE PLOT
# -------------------------------

model.fit(X, y)
importances = model.feature_importances_

feature_importance_df = pd.DataFrame({
    "Feature": X.columns,
    "Importance": importances
}).sort_values(by="Importance", ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(data=feature_importance_df, x="Importance", y="Feature")
plt.title("Feature Importance (Random Forest)")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.show()


# -------------------------------
# 3. PCA ON SELECTED FEATURES
# -------------------------------

print("\nApplying PCA...\n")

pca = PCA()
X_pca = pca.fit_transform(X_rfe)

explained_variance = pca.explained_variance_ratio_

print("Explained Variance Ratio:\n", explained_variance)


# -------------------------------
# 4. SCREE PLOT
# -------------------------------

plt.figure(figsize=(8,5))
plt.plot(range(1, len(explained_variance)+1), explained_variance, marker='o')
plt.title("Scree Plot")
plt.xlabel("Principal Components")
plt.ylabel("Variance Ratio")
plt.grid()
plt.show()


# -------------------------------
# 5. CUMULATIVE VARIANCE
# -------------------------------

cumulative_variance = np.cumsum(explained_variance)

plt.figure(figsize=(8,5))
plt.plot(range(1, len(cumulative_variance)+1), cumulative_variance, marker='o')
plt.title("Cumulative Explained Variance")
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Variance")
plt.grid()
plt.show()


# -------------------------------
# 6. FINAL PCA (95% VARIANCE)
# -------------------------------

print("\nReducing dimensions (retain 95% variance)...\n")

pca_final = PCA(n_components=0.95)
X_reduced = pca_final.fit_transform(X_rfe)

print("Original Shape:", X.shape)
print("After Feature Selection:", X_rfe.shape)
print("Final Reduced Shape:", X_reduced.shape)


# -------------------------------
# 7. FINAL DATAFRAME
# -------------------------------

X_reduced_df = pd.DataFrame(
    X_reduced,
    columns=[f"PC{i+1}" for i in range(X_reduced.shape[1])]
)

print("\nReduced Dataset Preview:")
print(X_reduced_df.head())


# -------------------------------
# 8. 2D VISUALIZATION
# -------------------------------

if X_reduced.shape[1] >= 2:
    plt.figure(figsize=(6,5))
    plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y, cmap='viridis', alpha=0.7)
    plt.xlabel("PC1")
    plt.ylabel("PC2")
    plt.title("PCA Projection")
    plt.colorbar(label="Price Movement")
    plt.show()


print("\n✅ Week 7 Completed: Top Feature Selection + PCA")
```
